## **Linear Regression for Africa**

```python
X = df['Year']
X = np.array(X).reshape(-1,1)
X.shape
```

```python
y = df['Africa']
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, train_size = 0.8, random_state = 88)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_train)
plt.figure(figsize = (15,5))
plt.scatter(X_train, y_train)
plt.plot(X_train, y_pred, color = 'b')
plt.title('Expenditure for Africa (Training set Result')
plt.xlabel('Year')
plt.ylabel('Expenditure')
plt.show()
```

```python
print('Coefficient of Determination:', model.score(X_train, y_train))
print('Intercept:', model.intercept_)
print('Slope:', model.coef_)
```

```python
x_2022 = np.array([2022]).reshape(-1,1)
Africa_pre = model.predict(x_2022)
print('Prediction for 2022:', model.predict(x_2022))
```
